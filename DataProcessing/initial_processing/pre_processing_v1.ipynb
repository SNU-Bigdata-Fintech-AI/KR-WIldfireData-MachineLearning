```python
# 원본데이터에 파생변수들 뽑아내고 수치형 자료 로그화
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import unicodedata
import os

# =========================
# 경로 설정
# =========================
INPUT_PATH = 'wildfire_2019_2023_combine.csv'
OUTPUT_PREP_PATH = os.path.join(os.path.dirname(INPUT_PATH), 'wildfire_preprocessed_V1.csv')
HIST_BINS = 20

plt.rcParams['axes.unicode_minus'] = False

# =========================
# 사용할 컬럼(실제 CSV는 대문자 컬럼)
# =========================
use_cols = [
    'IGTN_HTSRC_NM', 'CNTR_GRNDS_DSTNC', 'FRSTN_GRNDS_DSTNC',
    'HR_UNIT_ARTMP', 'HR_UNIT_HUM', 'HR_UNIT_WSPD_INFO',
    'CTPV_NM', 'CMBS_EXPOBJ_LCLSF_NM', 'FND_IGTN_PSTN_NM',
    'FND_FIRE_SE_NM', 'BGNN_POTFR_DT', 'DSPT_REQ_HR',
    'WNDRCT_BRNG', 'GRNDS_ARVL_DT', 'FIRE_SUPESN_HR',
    'RCPT_DT', 'DSPT_DT'
]

# 파일 읽기 (존재 컬럼만)
head_cols = pd.read_csv(INPUT_PATH, nrows=0).columns.str.upper().tolist()
existing_cols = [c for c in use_cols if c in head_cols]
df = pd.read_csv(INPUT_PATH, usecols=existing_cols, low_memory=False)
df.columns = [c.upper() for c in df.columns]  # 대문자 통일

# =========================
# (0) 'none'(대소문자 무관) 포함 행 제거
# =========================
mask_none_any = df.apply(lambda col: col.astype(str).str.lower().eq('none')).any(axis=1)
df = df[~mask_none_any].copy()

# =========================
# 유사 표기 정규화 유틸
# =========================
def normalize_basic(s: object) -> object:
    if pd.isna(s):
        return s
    s = str(s)
    s = unicodedata.normalize("NFKC", s)
    s = s.strip()
    s = re.sub(r"\s+", " ", s)
    return s

text_cols = ['IGTN_HTSRC_NM','FND_IGTN_PSTN_NM','FND_FIRE_SE_NM','CMBS_EXPOBJ_LCLSF_NM','CTPV_NM','WNDRCT_BRNG']
for c in text_cols:
    if c in df.columns:
        df[c] = df[c].apply(normalize_basic)

# =========================
# (2) HR_UNIT_ARTMP: -20~40
# =========================
if 'HR_UNIT_ARTMP' in df.columns:
    df['HR_UNIT_ARTMP'] = pd.to_numeric(df['HR_UNIT_ARTMP'], errors='coerce')
    df = df[df['HR_UNIT_ARTMP'].between(-20, 40, inclusive='both')].copy()

# =========================
# (3) HR_UNIT_HUM: 수치 변환
# =========================
if 'HR_UNIT_HUM' in df.columns:
    df['HR_UNIT_HUM'] = pd.to_numeric(df['HR_UNIT_HUM'], errors='coerce')

# =========================
# (4) HR_UNIT_WSPD_INFO → WSPD_BIN(1~5)
# =========================
def to_wspd_bin(x: object) -> float:
    if pd.isna(x):
        return np.nan
    s = normalize_basic(x).lower().replace(" ", "")
    s = s.replace("m/s","").replace("mps","").replace("～","~").replace("–","-")
    m = re.fullmatch(r"(\d+)[~-](\d+)", s)  # 범위
    if m:
        lo, hi = int(m.group(1)), int(m.group(2))
        if hi <= 4: return 1
        elif lo >= 5 and hi <= 8: return 2
        elif lo >= 9 and hi <= 12: return 3
        elif lo >= 13 and hi <= 17: return 4
        else: return 5
    m = re.fullmatch(r"(\d+)(이상|>=?)?", s)  # 단일/이상
    if m:
        v = int(m.group(1)); ge = m.group(2) is not None
        if ge: return 5 if v >= 18 else 4 if v >= 13 else 3 if v >= 9 else 2 if v >= 5 else 1
        return 1 if v <= 4 else 2 if v <= 8 else 3 if v <= 12 else 4 if v <= 17 else 5
    return np.nan

if 'HR_UNIT_WSPD_INFO' in df.columns:
    df['WSPD_BIN'] = df['HR_UNIT_WSPD_INFO'].map(to_wspd_bin)

# =========================
# (5) 날짜 파싱 + MONTH 생성(보강)
# =========================
def to_datetime_2step(s: pd.Series, fmt="%Y%m%d%H%M%S") -> pd.Series:
    # 1) 정형 포맷 시도 → 2) 실패분 유연 파싱
    s_str = s.astype(str)
    dt1 = pd.to_datetime(s_str, format=fmt, errors='coerce')
    need2 = dt1.isna()
    if need2.any():
        dt2 = pd.to_datetime(s_str[need2], errors='coerce', infer_datetime_format=True)
        dt1.loc[need2] = dt2
    return dt1

date_cols = ['RCPT_DT','DSPT_DT','GRNDS_ARVL_DT','BGNN_POTFR_DT']
for col in date_cols:
    if col in df.columns:
        df[col] = to_datetime_2step(df[col])

# MONTH: 우선 GRNDS_ARVL_DT, 없으면 RCPT_DT로 대체
if 'GRNDS_ARVL_DT' in df.columns:
    df['MONTH'] = df['GRNDS_ARVL_DT'].dt.month
    if df['MONTH'].isna().any() and 'RCPT_DT' in df.columns:
        df.loc[df['MONTH'].isna(), 'MONTH'] = df.loc[df['MONTH'].isna(), 'RCPT_DT'].dt.month
elif 'RCPT_DT' in df.columns:
    df['MONTH'] = df['RCPT_DT'].dt.month

# =========================
# (6) 범주형 인코딩
# =========================
kor2eng = {
    "서울특별시": "seoul", "부산광역시": "busan", "대구광역시": "daegu",
    "인천광역시": "incheon", "광주광역시": "gwangju", "대전광역시": "daejeon",
    "울산광역시": "ulsan", "경기도": "gyeonggi", "강원특별자치도": "gangwon",
    "충청북도": "chungbuk", "충청남도": "chungnam", "전북특별자치도": "jeonbuk",
    "세종특별자치시": "sejong", "전라남도": "jeonnam", "경상북도": "gyeongbuk",
    "경상남도": "gyeongnam", "제주특별자치도": "jeju",
    "남": "south", "북": "north", "동": "east", "서": "west",
    "공유림": "public", "국유림": "national", "사유림": "private",
    "미상": "unknown", "산아래": "under", "산정상": "top", "산중턱": "middle", "평지": "plain",
}

# 예외: CMBS_EXPOBJ_LCLSF_NM → dummy_na=True
if 'CMBS_EXPOBJ_LCLSF_NM' in df.columns:
    df['CMBS_EXPOBJ_LCLSF_NM'] = df['CMBS_EXPOBJ_LCLSF_NM'].replace(kor2eng)
    df = pd.get_dummies(df, columns=['CMBS_EXPOBJ_LCLSF_NM'], dummy_na=True, dtype=int)

other_cat_cols = [c for c in ['IGTN_HTSRC_NM','CTPV_NM','FND_IGTN_PSTN_NM','FND_FIRE_SE_NM','WNDRCT_BRNG'] if c in df.columns]
if other_cat_cols:
    for col in other_cat_cols:
        df[col] = df[col].replace(kor2eng)
        if df[col].isna().any():
            mode_val = df[col].mode(dropna=True)
            if len(mode_val) > 0:
                df[col] = df[col].fillna(mode_val.iloc[0])
    df = pd.get_dummies(df, columns=other_cat_cols, dummy_na=False, dtype=int)

# =========================
# (7) 결측치 처리
# =========================
if 'WSPD_BIN' in df.columns and df['WSPD_BIN'].isna().any():
    mode_val = df['WSPD_BIN'].mode(dropna=True)
    if len(mode_val) > 0:
        df['WSPD_BIN'] = df['WSPD_BIN'].fillna(mode_val.iloc[0])
    df['WSPD_BIN'] = df['WSPD_BIN'].round().astype('Int64')

for col in df.columns:
    if col in ['WSPD_BIN']:
        continue
    if pd.api.types.is_numeric_dtype(df[col]):
        if df[col].isna().any():
            df[col] = df[col].fillna(df[col].mean())

# =========================
# (8) 파생변수: df 내부에 직접 생성
# =========================
def _sec_nonneg(a, b):
    if a is None or b is None or (a is pd.NaT) or (b is pd.NaT):
        return np.nan
    d = (a - b).dt.total_seconds()
    return d.where(d >= 0, np.nan)

# 8-1) dispatch_time_diff: 신고~출동(초)
if {'RCPT_DT','DSPT_DT'}.issubset(df.columns):
    df['dispatch_time_diff'] = _sec_nonneg(df['DSPT_DT'], df['RCPT_DT'])
elif 'DSPT_REQ_HR' in df.columns:
    df['dispatch_time_diff'] = pd.to_numeric(df['DSPT_REQ_HR'], errors='coerce').where(lambda s: s >= 0, np.nan)
else:
    df['dispatch_time_diff'] = np.nan

# 8-2) arrival_time_diff: 신고~도착(초)
if {'RCPT_DT','GRNDS_ARVL_DT'}.issubset(df.columns):
    df['arrival_time_diff'] = _sec_nonneg(df['GRNDS_ARVL_DT'], df['RCPT_DT'])
else:
    df['arrival_time_diff'] = np.nan

# 8-3) golden_time_under_50min (3000초)
df['golden_time_under_50min'] = (df['DSPT_REQ_HR'] <= 3000).astype(int)

# 8-4) is_night (20:00~06:59) / month_rcpt
if 'RCPT_DT' in df.columns:
    hour = df['RCPT_DT'].dt.hour
elif 'DSPT_DT' in df.columns:
    hour = df['DSPT_DT'].dt.hour
else:
    hour = pd.Series(np.nan, index=df.index)
df['is_night'] = ((hour >= 20) | (hour <= 6)).astype('Int8')

if 'RCPT_DT' in df.columns:
    df['month_rcpt'] = df['RCPT_DT'].dt.month
else:
    df['month_rcpt'] = np.nan

# =========================
# (9) 로그 변환 & 원본 삭제
# =========================
log_vars = [
    'DSPT_REQ_HR',
    'FIRE_SUPESN_HR',
    'FRSTN_GRNDS_DSTNC',
    'CNTR_GRNDS_DSTNC',
    'arrival_time_diff',
    'dispatch_time_diff'
]

for col in log_vars:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col] = df[col].where(df[col] >= 0, np.nan).apply(np.log1p)


# =========================
# (마지막) 원본 날짜/풍속 문자열 삭제
# =========================
df.drop(columns=['GRNDS_ARVL_DT','BGNN_POTFR_DT','HR_UNIT_WSPD_INFO','RCPT_DT','DSPT_DT'], inplace=True, errors='ignore')

# =========================
# 전처리 CSV 저장
# =========================
df.to_csv(OUTPUT_PREP_PATH, index=False, encoding='utf-8-sig')

# =========================
# 화면에 기본 통계 & 히스토그램
# =========================
numeric_cols = df.select_dtypes(include=[np.number]).columns
print("=== DESCRIBE (NUMERIC) ===")
print(df[numeric_cols].describe())

for col in numeric_cols:
    try:
        df[col].hist(bins=HIST_BINS)
        plt.title(f'{col} Histogram')
        plt.xlabel(col); plt.ylabel('Frequency'); plt.grid(False)
        plt.show()
    except Exception:
        pass

print(f"\n✅ Preprocessed CSV saved to: {OUTPUT_PREP_PATH}")
```
