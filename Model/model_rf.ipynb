```python
# -*- coding: utf-8 -*-
"""
Leakage check & feature-importance comparison for wildfire suppression-time model.

- Compares performance with vs. without ARRIVAL_DELAY_MIN_LOG
- Trains RandomForestRegressor
- Prints RMSE/MAE/R2 and draws top-20 feature importances for both settings
- Also reports 5-fold CV RMSE for robustness
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# -----------------------------
# Config
# -----------------------------
DATA_PATH = "wildfire_preprocessed_3_logs.csv"
TARGET = "FIRE_SUPESN_HR"
LEAKY_FEATURE = "ARRIVAL_DELAY_MIN_LOG"
TEST_SIZE = 0.2
RANDOM_STATE = 42
N_ESTIMATORS = 200
N_JOBS = -1
TOP_K = 20

# -----------------------------
# Utils
# -----------------------------
def evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    mae = mean_absolute_error(y_test, pred)
    r2 = r2_score(y_test, pred)
    return rmse, mae, r2

def cv_rmse(model, X, y, n_splits=5, random_state=RANDOM_STATE):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
    # cross_val_score returns negative MSE because greater_is_better=True by default
    scores = cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=kf, n_jobs=N_JOBS)
    rmse_scores = np.sqrt(-scores)
    return rmse_scores.mean(), rmse_scores.std()

def plot_feature_importance(model, X, title, top_k=TOP_K):
    importances = model.feature_importances_
    idx = np.argsort(importances)[::-1][:top_k]
    plt.figure(figsize=(9, 7))
    plt.barh(range(len(idx)), importances[idx][::-1], align='center')
    plt.yticks(range(len(idx)), [X.columns[i] for i in idx][::-1])
    plt.xlabel("Feature Importance")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# -----------------------------
# Load
# -----------------------------
df = pd.read_csv(DATA_PATH)
y = df[TARGET].copy()

# A: all variables
X_all = df.drop(columns=[TARGET]).copy()

# B: without suspected leaky feature
drop_cols = [c for c in [TARGET, LEAKY_FEATURE] if c in df.columns]
X_no_delay = df.drop(columns=drop_cols).copy()

# -----------------------------
# Hold-out split
# -----------------------------
Xa_tr, Xa_te, ya_tr, ya_te = train_test_split(X_all, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)
Xb_tr, Xb_te, yb_tr, yb_te = train_test_split(X_no_delay, y, test_size=TEST_SIZE, random_state=RANDOM_STATE)

# -----------------------------
# Models
# -----------------------------
rf_a = RandomForestRegressor(n_estimators=N_ESTIMATORS, random_state=RANDOM_STATE, n_jobs=N_JOBS)
rf_b = RandomForestRegressor(n_estimators=N_ESTIMATORS, random_state=RANDOM_STATE, n_jobs=N_JOBS)

# -----------------------------
# Evaluate (hold-out)
# -----------------------------
rmse_a, mae_a, r2_a = evaluate(rf_a, Xa_tr, Xa_te, ya_tr, ya_te)
rmse_b, mae_b, r2_b = evaluate(rf_b, Xb_tr, Xb_te, yb_tr, yb_te)

comparison = pd.DataFrame({
    "Feature_Set": ["All variables", "Without ARRIVAL_DELAY_MIN_LOG"],
    "RMSE": [rmse_a, rmse_b],
    "MAE": [mae_a, mae_b],
    "R2": [r2_a, r2_b],
})

# -----------------------------
# 5-fold CV RMSE (robustness)
# -----------------------------
cv_rmse_a_mean, cv_rmse_a_std = cv_rmse(rf_a, X_all, y, n_splits=5)
cv_rmse_b_mean, cv_rmse_b_std = cv_rmse(rf_b, X_no_delay, y, n_splits=5)

cv_table = pd.DataFrame({
    "Feature_Set": ["All variables", "Without ARRIVAL_DELAY_MIN_LOG"],
    "CV_RMSE_Mean": [cv_rmse_a_mean, cv_rmse_b_mean],
    "CV_RMSE_Std": [cv_rmse_a_std, cv_rmse_b_std],
})

# -----------------------------
# Print results
# -----------------------------
print("\n=== Hold-out Performance ===")
print(comparison.to_string(index=False))
print("\n=== 5-Fold CV RMSE ===")
print(cv_table.to_string(index=False))

# -----------------------------
# Plot top-k feature importance
# -----------------------------
# Refit for plotting to ensure models are trained
rf_a.fit(Xa_tr, ya_tr)
rf_b.fit(Xb_tr, yb_tr)

plot_feature_importance(rf_a, Xa_tr, "Feature Importance (All variables)")
plot_feature_importance(rf_b, Xb_tr, "Feature Importance (Without ARRIVAL_DELAY_MIN_LOG)")

# -----------------------------
# Save utility as a reusable script
# -----------------------------
script_code = """
print('Hello from wildfire leakage check script!')
"""
with open("wildfire_leakage_check.py", "w", encoding="utf-8") as f:
    f.write(script_code)

```
